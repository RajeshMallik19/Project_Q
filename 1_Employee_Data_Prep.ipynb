```python
import pandas as pd

# Load the CSV file you just uploaded
file_path = '/content/WA_Fn-UseC_-HR-Employee-Attrition.csv'
df = pd.read_csv(file_path)

# --- Verification Step ---
# Display the first 5 rows to ensure it loaded correctly
print("Data loaded successfully. Here are the first 5 rows:")
df.head()
```

```python
# Get a concise summary of the dataframe
# This shows column names, non-null counts, and data types (Dtype)
print("\nDataset Information:")
df.info()
```

```python
# Check how many unique values are in these columns
print(f"\nUnique values in 'EmployeeCount': {df['EmployeeCount'].nunique()}")
print(f"Unique values in 'StandardHours': {df['StandardHours'].nunique()}")
print(f"Unique values in 'Over18': {df['Over18'].nunique()}")

# Drop these constant columns, plus the employee number which is just an ID
df_cleaned = df.drop(columns=['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'])

# --- Verification Step ---
# Display the shape of the dataframe before and after
print(f"\nOriginal shape (rows, columns): {df.shape}")
print(f"Shape after cleaning (rows, columns): {df_cleaned.shape}")
```

```python
# We will work with the cleaned dataframe from the previous step
# The variable was named df_cleaned

# Separate the target variable (what we want to predict)
target = df_cleaned['Attrition']

# Separate the feature variables (the inputs to the model)
features = df_cleaned.drop(columns=['Attrition'])

# --- Encode the Target Variable ---
# We use LabelEncoder to turn 'No' and 'Yes' into 0 and 1
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
target_encoded = le.fit_transform(target)

# --- Verification Step ---
# Check the first 5 and last 5 values to see the change
# Original: ['Yes', 'No', 'Yes', 'No', 'No']
# Encoded: [1, 0, 1, 0, 0]
print("Original Attrition labels:", target.head().values)
print("Encoded Attrition labels: ", target_encoded[:5])

# --- One-Hot Encode the Feature Variables ---
# This will find all columns with text data and convert them into numerical columns
features_encoded = pd.get_dummies(features)

# --- Verification Step ---
# Display the first 5 rows of the new, fully-encoded dataframe
print("\nFeatures after one-hot encoding:")
features_encoded.head()

# Check the shape of our final encoded features dataframe
print(f"\nOriginal number of feature columns: {features.shape[1]}")
print(f"Number of feature columns after encoding: {features_encoded.shape[1]}")
```

```python
# Create a new dataframe for our final, prepped data
# This combines the encoded features and the encoded target
final_df = features_encoded.copy()
final_df['Attrition'] = target_encoded

# Save this dataframe to a new CSV file
# The index=False part is important to avoid saving an extra column
final_df.to_csv('employee_attrition_prepped.csv', index=False)

print("Prepped data saved to 'employee_attrition_prepped.csv'")
print("You can now start a new notebook for model training.")
```
