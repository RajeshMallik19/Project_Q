```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ===================================================================
# Part 1: Manually Enter Your Final Results for the Student Dataset
# ===================================================================
# These are the F1-Scores for class '1' (at-risk) from all your work
results = {
    'Model': [
        'Random Forest (Baseline)',
        'XGBoost (Baseline)',
        'BiLSTM-CNN (Classical)',
        'Quantum Hybrid'
    ],
    'F1-Score': [
        0.75,  # From notebook Student_2
        0.79,  # From notebook Student_2
        0.81,  # From notebook Student_3
        0.75   # From notebook Student_4
    ]
}
results_df = pd.DataFrame(results).sort_values('F1-Score', ascending=False)

print("--- Final Results Summary for Student At-Risk Prediction ---")
print(results_df)

# ===================================================================
# Part 2: Generate the Final Comparison Plot for Your Paper
# ===================================================================
print("\n--- Generating Final Plot ---")
plt.figure(figsize=(12, 7))
sns.set_theme(style="whitegrid")

# Create the bar plot
barplot = sns.barplot(
    x='F1-Score',
    y='Model',
    data=results_df,
    palette='viridis'
)

# Add data labels to each bar for clarity
for index, row in results_df.iterrows():
    barplot.text(row['F1-Score'] / 2, index, f"{row['F1-Score']:.2f}",
                 color='white', ha="center", va="center", weight='bold')

plt.title('Final Model Performance Comparison (F1-Score for At-Risk Students)', fontsize=16)
plt.xlabel('F1-Score', fontsize=12)
plt.ylabel('Model', fontsize=12)
plt.xlim(0, 1.0)
plt.show()
```
