```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ===================================================================
# Part 1: Manually Enter Your Final, Real Results
# ===================================================================
# These are the F1-Scores for class '1' (Attrition) from all your work
results = {
    'Model': [
        'Random Forest (Baseline)',
        'XGBoost (Baseline)',
        'BiLSTM-CNN (Classical)',
        'Quantum Hybrid (Proof-of-Concept)'
    ],
    'F1-Score': [
        0.35,  # From notebook 2
        0.41,  # From notebook 2
        0.48,  # From notebook 3
        0.28   # From notebook 4
    ]
}
results_df = pd.DataFrame(results).sort_values('F1-Score', ascending=False)

print("Final Results Summary:")
print(results_df)

# ===================================================================
# Part 2: Generate the Final Comparison Plot for Your Paper
# ===================================================================
plt.figure(figsize=(12, 7))
sns.barplot(x='F1-Score', y='Model', data=results_df, palette='viridis')

# Add labels to the bars
for index, row in results_df.iterrows():
    plt.text(row['F1-Score'] + 0.01, index, f"{row['F1-Score']:.2f}", color='black', ha="left", va="center")

plt.title('Final Model Performance Comparison (F1-Score for Attrition)', fontsize=16)
plt.xlabel('F1-Score', fontsize=12)
plt.ylabel('Model', fontsize=12)
plt.xlim(0, 0.6)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
