```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv('xAPI-Edu-Data.csv')

print("--- Original Data Info ---")
df.info()
print("\n--- First 5 Rows ---")
print(df.head())

# --- Target Variable Engineering ---
# The 'Class' column has three categories: L, M, H.
# We will define our prediction problem as identifying at-risk students.
# We'll map 'L' (Low-performing) to 1 (at-risk) and 'M' and 'H' to 0 (not at-risk).
df['at_risk'] = df['Class'].apply(lambda x: 1 if x == 'L' else 0)

# We can now drop the original 'Class' column
df = df.drop(columns=['Class'])

print("\n--- Target Variable 'at_risk' Created ---")
print(df['at_risk'].value_counts())


# --- Feature Engineering ---
# We will one-hot encode all categorical columns.
# First, identify which columns are categorical (dtype 'object')
categorical_cols = df.select_dtypes(include=['object']).columns
print(f"\nCategorical columns to be one-hot encoded: {list(categorical_cols)}")

# Use pandas get_dummies for easy one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print("\n--- Data After One-Hot Encoding ---")
print(df_encoded.head())


# --- Final Step: Save the Prepped Data ---
# Save this fully prepped dataframe to a new CSV file for the next notebooks.
df_encoded.to_csv('student_dropout_prepped.csv', index=False)

print("\n-------------------------------------------------")
print("Preprocessing complete!")
print("Clean data saved to 'student_dropout_prepped.csv'")
print(f"Final dataset has {df_encoded.shape[0]} rows and {df_encoded.shape[1]} columns.")
```
